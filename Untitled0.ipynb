```python
def calculate_grade(percentage):
    if percentage >= 90:
        return "A"
    elif percentage >= 75:
        return "B"
    elif percentage >= 60:
        return "C"
    elif percentage >= 50:
        return "D"
    else:
        return "F"

def main():
    # PART 1: Welcome Screen
    print("=====================================")
    print("       STUDENT PROGRESS TRACKER      ")
    print("=====================================")

    # PART 2: Enter Subject Marks
    student_name = input("Enter student name: ")
    roll_no = input("Enter roll number: ")

    print("
Enter marks for 5 subjects (0 - 100):")
    marks = []
    for i in range(1, 6):
        m = float(input(f"Subject {i}: "))
        marks.append(m)

    # PART 3: Grade Calculation
    total = sum(marks)
    percentage = total / len(marks)
    grade = calculate_grade(percentage)

    # PART 4: Pass / Fail Logic
    # Pass if percentage >= 40 and no subject below 35
    passed = percentage >= 40 and all(m >= 35 for m in marks)

    # PART 5: Display Progress Report
    print("
=========== PROGRESS REPORT =========")
    print(f"Name      : {student_name}")
    print(f"Roll No   : {roll_no}")
    print("-------------------------------------")
    for i, m in enumerate(marks, start=1):
        print(f"Subject {i} : {m:.2f}")
    print("-------------------------------------")
    print(f"Total     : {total:.2f} / 500.00")
    print(f"Percent   : {percentage:.2f}%")
    print(f"Grade     : {grade}")
    print(f"Result    : {'PASS' if passed else 'FAIL'}")
    print("=====================================")

if __name__ == "__main__":
    main()
```
